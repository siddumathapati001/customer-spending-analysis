## 2. Data Cleaning & Preprocessing

```python
# Creating working copy
df_clean = df.copy()
```

```python
df_clean.isna().sum().sort_values(ascending = False)
```

```python
# --- Verification: Income distribution ---
df_clean["Income"].skew()
```

```python
# Handling Missing Values
## Income has missing values - since it is numeric column 
## Impute with median as it is better than mean (outliers and Skeweness)

income_median = df_clean['Income'].median()
df_clean['Income'] = df_clean['Income'].fillna(income_median)
```

```python
# Setting Dates in proper format
df_clean['Dt_Customer'] = pd.to_datetime(df_clean['Dt_Customer'], format="mixed")
```

```python
# Creating Tenure column using Dt_Customer
df_clean['Customer_Tenure_Days'] = (
    pd.to_datetime('today') - df_clean['Dt_Customer']
).dt.days
```

```python
# Droping ID column as it has no use in Analysis
df_clean.drop(columns=['ID'], inplace=True)
```

```python
# Checking for Duplicates
df_clean.duplicated().sum()
```

```python
# Droping Duplicates
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.info()
```

## 3. Feature Engineering

```python
# Craeting Total_Spending Column
spending_cols = [
    'MntWines', 'MntFruits',
    'MntMeatProducts', 'MntFishProducts', 
    'MntSweetProducts', 'MntGoldProds'
]

df_clean['Total_Spending'] = df_clean[spending_cols].sum(axis=1)
```

```python
df_clean[spending_cols + ['Total_Spending']].head()
```

```python
# Creating Family_Size Column
df_clean['Family_Size'] = df_clean['Kidhome'] + df_clean['Teenhome'] + 1
```

```python
# Converting Education into Orderd Categories
education_order = {
    'Basic' : 0,
    '2n Cycle' : 1,
    'Graduation' : 2,
    'Master' : 3,
    'PhD' : 4
}
df_clean['Education_Level'] = df_clean['Education'].map(education_order)
```

## 5. Statistical Analysis

#### Stastistical Significance between Income and Total Spending
- H0 (Null Hypothesis) : There is no linear relationship between Income and Total_Spending.
- H1 (Alternative Hypothesis) : There is a linear relationship between Income and Total_Spending.
- Both are Numerical Columns --> Pearson Correlation test

```python
from scipy import stats
# Pearson Test
corr, p_value = stats.pearsonr(df_clean['Income'], df_clean['Total_Spending'])

print(f"Pearson correlation: {corr:.3f}")
print(f"P-value: {p_value:.5f}")

if p_value < 0.05:
    print('Reject Null Hypothesis')
else:
    print('Fail to Reject Null Hypothesis')
```

Income shows a statistically significant positive linear relationship with total spending. However, the presence of high-income customers with low spending indicates that income alone does not fully explain customer purchasing behavior.

#### Statistical Significance between Education Level and Total Spending
- HO (Null Hypothesis) : Average total spending is the same across all education levels.
- H1 (Alternate Hypothesis) : At least one education level has a different average total spending.
- Total_Spending is numerical and Educattion is categorical with more than 2 groups --> ANOVA 

```python
# Creating groups of different Education Levels and Spending 
education_levels = df_clean["Education"].unique()

groups = [
    df_clean[df_clean["Education"] == edu]["Total_Spending"]
    for edu in education_levels
]
# ANOVA test
f_stat, p_value = stats.f_oneway(*groups)

print(f"F_statistics : {f_stat:.4f}")
print(f"P_value : {p_value:.5f}")

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

The ANOVA test indicates a statistically significant difference in average total spending across education levels (p < 0.05). This suggests that customer spending behavior varies by education level, although further analysis is required to identify which specific groups differ. 

This statistical result aligns with exploratory analysis, which showed variation in median spending across education groups, though with overlapping distributions.



#### Statistica Significance between Marital Status and Spending
- H0 (Null Hypothesis) : Average total spending is same across different marital status
- H1 (Alternate Hypothesis) : At least one marital status has a different average total spending

```python
# Creating groups for different marital status
marital_status_groups = [df_clean[df_clean['Marital_Status'] == status]['Total_Spending'] for status in df_clean['Marital_Status'].unique()]

# Oneway ANOVA test
f_stat, p_value = stats.f_oneway(*marital_status_groups)

#interpretation
print(f"F_statistics : {f_stat:.4f}")
print(f"P_value : {p_value:.5f}")

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

The ANOVA test did not find a statistically significant difference in mean total spending across marital status groups (p > 0.05). This suggests that marital status alone does not have a strong standalone association with customer spending behavior.

Although marital status does not show a significant standalone effect, its influence on spending may be indirect through factors such as household size or number of dependents.

# Overall Conclusion
Overall, customer spending behavior is influenced more strongly by income and education-related factors than by marital status alone. Revenue is driven primarily by a few key product categories, suggesting that targeted engagement strategies focused on high-value segments and premium products could yield meaningful business impact.
